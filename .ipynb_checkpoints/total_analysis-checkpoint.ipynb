```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
def get_ref():
    
    clash_df = pd.read_csv("analysis/reference-clashes.csv", index_col = 0)
    cor_df = pd.read_csv("analysis/reference-correlation.csv", index_col = 0)
    ref_df = clash_df.join(cor_df["correlation"])
    return ref_df
```

```python
def make_tops_dicts(run, sorter, colorcol="correlation"):    
    
    tops = [50, 25, 10, 5, 1]
    df_list = make_df_list(run)
    ref_df = get_ref()
    annot_dict = {}
    color_dict = {}
    
    for df in df_list:

        if sorter == "correlation":
            df = df.sort_values([sorter], ascending=False)
        else:
            df = df.sort_values([sorter])    

        compl = str(df.index[0][:4])
        cor = df.loc[:,"correlation"]
        clash = df.loc[:,"atomic-clashes"]

        annot_dict[compl] = {}
        color_dict[compl] = {}
        
        if colorcol == "correlation":
            for top in tops:
                mean_cor = round(cor.iloc[0:top].mean(), 4)
                color_dict[compl][top] = mean_cor

                mean_clash = round(clash.iloc[0:top].mean(), 1)
                annot_dict[compl][top] = mean_clash

            annot_dict[compl]["ref"] = ref_df.loc[compl, "clashes"]
            color_dict[compl]["ref"] = ref_df.loc[compl, "correlation"]
        else:
            for top in tops:
                mean_cor = round(cor.iloc[0:top].mean(), 4)
                annot_dict[compl][top] = mean_cor

                mean_clash = round(clash.iloc[0:top].mean(), 1)
                color_dict[compl][top] = mean_clash

            color_dict[compl]["ref"] = ref_df.loc[compl, "clashes"]
            annot_dict[compl]["ref"] = ref_df.loc[compl, "correlation"]

    annot_df = pd.DataFrame.from_dict(annot_dict)
    color_df = pd.DataFrame.from_dict(color_dict)
    
    for index in color_df.columns:
        ref = ref_df.loc[index, colorcol]
        color_df[index] = color_df.loc[:,index].div(ref).mul(100)
    if colorcol == "correlation":
        color_df = color_df.sub(100)
    
    return color_df, annot_df
```

```python
def make_scatters(run):
    ref_df = get_ref()
    df_list = make_df_list(run)
    lst = [0, 3, 6, 9, 12, 15]
    clash = []
    cor = []

    fig, ax = plt.subplots(nrows=6, ncols=3, figsize=(18,36))
    column = -1
    for i, df in enumerate(df_list):
        if i in lst:
            row = 0
            column += 1
        else:
            row += 1


        clash = []
        cor = []
        compl = df.index[0][:4]
        ref_cor = ref_df.loc[compl, "correlation"]
        ref_clash = ref_df.loc[compl, "clashes"]
        df["atomic-clashes"] = df["atomic-clashes"].div(ref_clash).mul(100)
        df["correlation"] = df["correlation"].div(ref_cor).mul(100)

        for struct in df.index:
            clash.append(df.loc[struct, "atomic-clashes"])
            cor.append(df.loc[struct, "correlation"])

        plot = sns.scatterplot(clash, cor, ax=ax[column, row])
        plot.set_title(compl)
        plot.set_xlabel("Remaining clashes %")
        plot.set_ylabel("Correlation to map relative to ref %")

```

```python
# divide the scatterplot in 9 boxes with 3 levels of correlation and 3 levels of clash removal

high = -0.2
medium = -1

good = 1
fair = 10



high_good = {}
high_fair = {}
high_poor = {}

medium_good = {}
medium_fair = {}
medium_poor = {}

low_good = {}
low_fair = {}
low_poor = {}

data_list = [high_good, medium_good, low_good, high_fair, medium_fair, low_fair, high_poor, medium_poor, low_poor]            
total_structs_dict = {}
for run in cor_dict:
    run_cor = cor_dict[run]
    run_clash = clash_dict[run]
    
    #init the counting
    high_good[run[13:]] = 0
    high_fair[run[13:]] = 0
    high_poor[run[13:]] = 0
    medium_good[run[13:]] = 0
    medium_fair[run[13:]] = 0
    medium_poor[run[13:]] = 0
    low_good[run[13:]] = 0
    low_fair[run[13:]] = 0
    low_poor[run[13:]] = 0
    
    total_structs = 0
    for struct in run_cor:
        cors = run_cor[struct]
        clashes = run_clash[struct]
        
        for i in range(len(cors)):
            cor = cors[i]
            clash = clashes[i]
            total_structs += 1
            
            if cor > high and clash < good:
                high_good[run[13:]] += 1

            elif cor > medium and clash < good:
                medium_good[run[13:]] += 1

            elif clash < good:
                low_good[run[13:]] += 1
                
            elif cor > high and clash < fair:
                high_fair[run[13:]] += 1

            elif cor > medium and clash < fair:
                medium_fair[run[13:]] += 1

            elif clash < fair:
                low_fair[run[13:]] += 1

            elif cor > high:
                high_poor[run[13:]] += 1

            elif cor > medium:
                medium_poor[run[13:]] += 1

            else:
                low_poor[run[13:]] += 1
                
    for index in data_list:
        index[run[13:]] = index[run[13:]] / total_structs * 100
        
    total_structs_dict[run[13:]] = total_structs

    print("\tGood\tFair\tPoor")    
    print("High\t{}\t{}\t{}".format(high_good[run[13:]], high_fair[run[13:]], high_poor[run[13:]]))
    print("Medium\t{}\t{}\t{}".format(medium_good[run[13:]], medium_fair[run[13:]], medium_poor[run[13:]]))
    print("low\t{}\t{}\t{}\n".format(low_good[run[13:]], low_fair[run[13:]], low_poor[run[13:]]))

for struct in total_structs_dict:    
    print(struct, total_structs_dict[struct])
```

```python
def make_bar(data_list):
    
    fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(18,18)) 
#     plt.subplots_adjust(wspace=0.05, hspace=0.05)
    
    lst = [0, 3, 6, 9, 12, 15]
    column = -1
    
    for i, datas in enumerate(data_list):
        runs = []
        data = []
        
        if i in lst:
            row = 0
            column += 1
        else:
            row += 1
        
        for run in datas:
            data.append(datas[run])
            runs.append(run)

        plot = sns.barplot(x=runs, y=data, ax = ax[row, column])
#         plot.set_ylim((0,50))
        
        plot.set_xticklabels(plot.get_xticklabels(), rotation=45, ha="right", fontsize = "x-large")
        plot.set_ylim((0,30))
        
        if row < 2:
            plot.set_xticks([])
            
          
        if row == 0:
            if column == 0:
                plot.set_ylabel("% of all", fontsize ="x-large")
                
                plot.set_title("< 1% remaining clashes", fontsize = "x-large")
            
            if column == 1:
                plot.set_title("1% - 10% remaining clashes", fontsize = "x-large")
            
            if column == 2:
                plot.set_title("> l0% remainig clashes", fontsize = "x-large")
                
                plot2 = plot.twinx()
                plot2.set_yticks([])
                plot2.yaxis.labelpad = 25
                plot2.set_ylabel(" < 0.2% correlation loss", fontsize= "x-large", rotation = 270)
        
        if row == 1:
            if column == 0:
                
                plot.set_ylabel("% of all", fontsize = "x-large")
                
            if column == 2:
                plot2 = plot.twinx()
                plot2.set_yticks([])
                plot2.yaxis.labelpad = 25
                plot2.set_ylabel("0.2 - 1% correlation loss", fontsize= "x-large", rotation = 270)

             
        if row == 2:
            if column == 0:
                
                plot.set_ylabel("% of all", fontsize = "x-large")
                
            if column == 2:
                plot2 = plot.twinx()
                plot2.set_yticks([])
                plot2.yaxis.labelpad = 25
                plot2.set_ylabel("> 1% correlation loss", fontsize= "x-large", rotation = 270)
                
                                
            
data_list = [high_good, medium_good, low_good, high_fair, medium_fair, low_fair, high_poor, medium_poor, low_poor]            
make_bar(data_list)
#plt.savefig("barplot", dpi = 100)
```
